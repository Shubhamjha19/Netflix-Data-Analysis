# NETFLIX Data Analysis


```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the data of netflix
data = pd.read_csv('netflix.csv')
data
```

```python
## Droping the shows whos director are not listed:
data = data[data['director'] != 'Not Given']
data
```

```python
data['listed_in'].unique()
```

```python
## getting the number of occurence for a single category


data['listed_in'].value_counts()['Comedies']
```

```python
## getting all the number of occurence for multiple categories
data['listed_in'].value_counts().to_dict()
```

```python
## Finding all the data that are of Movie type


data[data['type'].str.contains('Movie',case=False)]
```

```python
# filtering out only the films that have been made in 2021

year_2021= data[data['release_year'] == 2021]
yr = year_2021.value_counts()
avg_10= yr.head(10)
avg_10
```

```python
data.head()
```

```python
data['listed_in'].unique()
```

```python
# Check out how much of each 4 categories(listed_in) are in percentage:
categories = ['Action','Comedies','Thrillers','Dramas']
categ ={}
for i in categories:
    categ[i] = data['listed_in'].str.contains(i).sum()
p = pd.Series(categ)
plt.figure(figsize=(6,6))
plt.pie(p,labels=p.index,autopct='%2f')
plt.title("Netflix type of movies inclusive of 4 types",color='k',size=15)
plt.show()
```

```python
req_year = [1993,2000,2006,2019,2021]

count ={}

for i in req_year:
    count[i] = data[data['release_year']== i]

df = pd.concat(count.values(),keys=count.keys())

yrs = df['release_year'].value_counts()

plt.figure(figsize = (6,6))
plt.pie(yrs,labels=yrs.index,autopct='%.2f%%')
plt.title('Netflix films released in particular years',size=15)
plt.show()
```

```python
# Distribution of movies,Tv Shows around years

title = data['type']
year= data['release_year']
plt.figure(figsize = (6,8))

plt.scatter(x=title,y= year)
plt.title('Scattered Plot showing distribution of types of films around years',size=12)
plt.show()
```

```python
# Viewing rate of growth using ploting
plt.plot(yrs)
plt.title('Growth of productions in Netflix around different years',size=15)
plt.xlabel('Years',size = 15)
plt.ylabel('Films Produced',size=15)
plt.grid = True
plt.show()
```

```python
sns.barplot(yrs)
plt.title('Growth of productions in Netflix around different years ',size=15)
plt.xlabel('Years',size = 15)
plt.ylabel('Films Produced',size=15)
plt.show()
```
